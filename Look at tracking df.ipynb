```python
from whotracksme.data import DataSource
```

```python
data = DataSource()
```

```python
trackers_df = data.trackers.df
trackers_df.head()
```

```python
fp_data = []
for month in trackers_df.month.unique():
    month_df = trackers_df[trackers_df.month == month]
    d = (f'{month}'[0:7], f'{sum(month_df.bad_qs > 0.1) / len(month_df):.1%}')
    fp_data.append(d)
    print(d)
```
